```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/fenago/pythonml/main/data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
%time
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
# Address Missing Values
df.info()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

string_columns = list(df.dtypes[df.dtypes == 'object'].index)

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
## Traget value to be labelled to numeric ( # Treat the target in binary classifciation)
# Treat the target in binary classification (make sure there are only 2 values)
from sklearn.preprocessing import LabelEncoder

# Instantiate the LabelEncoder
le = LabelEncoder()

# re
# Fit and transform the binary column
df['churn'] = le.fit_transform(df['churn'])

# Print the mapping
mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(mapping)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
## True training dataset, remaining validation dataset
df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=11)
y_train=df_train.churn.values
y_val=df_val.churn.values
del df_train['churn']
del df_val['churn']
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def calculate_mi(series):
    return mutual_info_score(series, df_train_full.churn)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')


display(df_mi.head())
display(df_mi.tail())
```

# Feature Engineering

```python
## One Hot Encoding
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient='records')
```

```python
train_dict[0]
```

```python
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
X_train.shape
```

```python
feature_names = dv.feature_names_
feature_names
```

```python
feature_names = dv.get_feature_names_out()
feature_names
```

# Machine Learning

```python
## Using Logistic Regression - Sigmoid function for Binary Classification model
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
model.predict_proba(X_val)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
churn = y_pred > 0.5
```

```python
## WHich percentage of customers are going to churn
(y_val == churn).mean()
```
